## a)
E, F, A, B, D, H, G, C

## b)

$P(A, B, C, D, E, F, G, H) = P(A|F) P(B) P(C|F,H) P(D|A,E,F) P(E) P(F) P(G|A,H)$


## c)
The markov blanket of node a has D, G, H as children, F as parents and E and B as parents of children.

The orginal DAG:

```python
import networkx as nx


%matplotlib inline

G = nx.DiGraph()
G.graph['dpi'] = 120
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "H"])
G.add_edges_from([("A", "B"), ("F", "D"), ("F", "C"), ("F", "A"), ("E", "D"), ("A", "H"), ("B", "H"), ("H", "G"), ("H", "C"), ("A", "G"), ("A", "D")])

nx.draw_planar(G, with_labels=True)
```

And now the moral graph that removes directed edges and addes edges between the markov blankets of all nodes

```python
import networkx as nx


%matplotlib inline

G = nx.Graph()
G.graph['dpi'] = 120
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "H"])
G.add_edges_from([("A", "B"), ("F", "D"), ("F", "C"), ("F", "A"), ("E", "D"), ("A", "H"), ("B", "H"), ("H", "G"), ("H", "C"), ("A", "G"), ("A", "D")])

# add edges
G.add_edges_from([("E", "F"), ("E", "A"), ("F", "H"), ("A", "B")])

nx.draw_planar(G, with_labels=True)
```
